```python
%matplotlib inline
import pymc as pm
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
data = np.loadtxt("data/mixture_data.csv", delimiter=",")
data[1:5,]
```

```python
plt.hist(data, bins=20, color="k", histtype="stepfilled", alpha=0.7)
plt.xlim(0,None)
#it seems there are two distrobutions within the data set? what are their parameters?
```

```python
#create a uniform variable that represents the probability of belonging to one of the clusters
p = pm.Uniform("p", 0, 1)
```

```python
#check our uniform distrobution to insure
samples = [p.random() for i in range(10000)]
plt.hist(samples, color="k", alpha=.7)
plt.xlim(0,None)
```

```python
#pm.Categorical??
assignment = pm.Categorical("assignment", [p, 1 - p], size=len(data))
print(assignment.value[:10], "...")
```

```python
#we can create two variables by specifying size=N
stds = pm.Uniform("stds", 0, 100, size=2)
taus = 1.0 / stds ** 2
centers = pm.Normal("centers", [120, 190], [0.01, 0.01], size=2)
```

```python
@pm.deterministic
def center_i(assignment=assignment, centers=centers):
    return centers[assignment]

@pm.deterministic
def tau_i(assignment=assignment, taus=taus):
    return taus[assignment]
```

```python
#add our observed data
observations = pm.Normal("obs", center_i, tau_i, value=data, observed=True)

#then create model class
model = pm.Model([p, assignment, observations, taus, centers])
```

```python
mcmc = pm.MCMC(model)
mcmc.sample(100000)
```

```python
plt.hist(mcmc.trace("centers")[:,0],bins=30,alpha=.7)
plt.hist(mcmc.trace("centers")[:,1],bins=30,alpha=.7)
plt.xlim(100,None)
```

```python
plt.hist(mcmc.trace("stds")[:,0],bins=30,alpha=.7)
plt.hist(mcmc.trace("stds")[:,1],bins=30,alpha=.7)
plt.xlim(0,None)
```

```python
plt.scatter(data, 1 - (mcmc.trace("assignment")[:]).mean(axis=0))
plt.ylim(-0.05, 1.05)
plt.xlim(35, 300)
```

```python
x = np.linspace(20, 300, 500)
posterior_center_means = (mcmc.trace("centers")[:]).mean(axis=0)
posterior_std_means = (mcmc.trace("stds")[:]).mean(axis=0)
posterior_p_mean = (mcmc.trace("p")[:]).mean()
```

```python
plt.hist(data, bins=20, histtype="step", normed=True, color="k",lw=2,label="Actual Data")
plt.plot(x, posterior_p_mean * 
         stats.norm.pdf(x, loc=posterior_center_means[0],scale=posterior_std_means[0]),color="k",label="Cluster0")
plt.fill_between(x, posterior_p_mean * 
                 stats.norm.pdf(x, loc=posterior_center_means[0],scale=posterior_std_means[0]),color="k",alpha=.3)

plt.plot(x, (1-posterior_p_mean) * 
         stats.norm.pdf(x, loc=posterior_center_means[1],scale=posterior_std_means[1]),color="g",label="Cluster1")
plt.fill_between(x, (1-posterior_p_mean) * 
                  stats.norm.pdf(x, loc=posterior_center_means[1],scale=posterior_std_means[1]),color="g",alpha=.3)
plt.grid(True)
plt.legend(loc="upper left")
```
